```python
import psycopg2
import csv

conn = psycopg2.connect(
    host="localhost",
    port="5432",
    dbname="phonebook",
    user="postgres",
    password="yernur06"
)
cursor = conn.cursor()

cursor.execute('''
    CREATE TABLE IF NOT EXISTS phonebook (
        id SERIAL PRIMARY KEY,
        name VARCHAR(100),
        phone VARCHAR(20)
    )
''')
conn.commit()

def insert_from_console():
    name = input("name: ")
    phone = input("phone: ")
    cursor.execute("INSERT INTO phonebook (name, phone) VALUES (%s, %s)", (name, phone))
    conn.commit()

def insert_from_csv():
    with open('C:/Users/User/Downloads/contacts.csv', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            cursor.execute("INSERT INTO phonebook (name, phone) VALUES (%s, %s)", (row['name'], row['phone']))
    conn.commit()

def update_user():
    name = input("name:")
    new_phone = input("new phone: ")
    cursor.execute("UPDATE phonebook SET phone = %s WHERE name = %s", (new_phone, name))
    conn.commit()

def query_users():
    pattern = input("pattern: ")
    cursor.execute("SELECT * FROM phonebook WHERE name ILIKE %s OR phone ILIKE %s", (f"%{pattern}%", f"%{pattern}%"))
    for row in cursor.fetchall():
        print(row)

def delete_user():
    key = input("key: ")
    cursor.execute("DELETE FROM phonebook WHERE name = %s OR phone = %s", (key, key))
    conn.commit()

while True:
    print("\n1 - insert by console\n2 - insert from csv\n3 - update\n4 - query users\n5 - delete\n0 - exit")
    choice = input()

    if choice == '1':
        insert_from_console()
    elif choice == '2':
        insert_from_csv()
    elif choice == '3':
        update_user()
    elif choice == '4':
        query_users()
    elif choice == '5':
        delete_user()
    elif choice == '0':
        break
    else:
        print("Error")

cursor.close()
conn.close()
```

```python
import random
import pygame
import psycopg2
pygame.init()

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)

size = (700, 500)
screen = pygame.display.set_mode(size)
pygame.display.set_caption("Snake")
clock = pygame.time.Clock()

conn = psycopg2.connect(
    host="localhost",
    port="5432",
    dbname="phonebook",
    user="postgres",
    password="yernur06"
)
cursor = conn.cursor()

username = input("Enter your username: ")
cursor.execute("SELECT id FROM users WHERE username = %s", (username,))
user = cursor.fetchone()

if user is None:
    cursor.execute("INSERT INTO users (username) VALUES (%s) RETURNING id", (username,))
    user_id = cursor.fetchone()[0]
    print(f"New user created with ID: {user_id}")
else:
    user_id = user[0]
    print(f"Welcome back! User ID: {user_id}")

rect_x = 350
rect_y = 250
rect_change_x = 15
rect_change_y = 15
right = True
left = up = down = False
snake = []
scope = 1
block = 'a'

foods = []
def spawn_food():
    return {
        "x": random.randint(0, 650),
        "y": random.randint(0, 350),
        "time_left": 5000,  # in milliseconds (5 seconds)
        "spawn_time": pygame.time.get_ticks()
    }
foods.append(spawn_food())

font = pygame.font.Font(None, 30)

done = False
while not done:
    screen.fill(BLACK)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_w and block != 'w':
                block = 's'
                right = left = down = False
                up = True
            elif event.key == pygame.K_s and block != 's':
                block = 'w'
                right = left = up = False
                down = True
            elif event.key == pygame.K_a and block != 'a':
                block = 'd'
                right = up = down = False
                left = True
            elif event.key == pygame.K_d and block != 'd':
                block = 'a'
                left = up = down = False
                right = True

    if right: rect_x += rect_change_x
    if left: rect_x -= rect_change_x
    if up: rect_y -= rect_change_y
    if down: rect_y += rect_change_y

    rect_x %= 700
    rect_y %= 500

    snake.append((rect_x, rect_y))
    snake = snake[-scope * 2:]

    if len(snake) != len(set(snake)):
        done = True

    for i, j in snake:
        pygame.draw.rect(screen, GREEN, [i, j, 15, 15])

    current_time = pygame.time.get_ticks()
    for food in foods[:]:
        elapsed = current_time - food["spawn_time"]
        time_left = max(0, food["time_left"] - elapsed)
        if time_left <= 0:
            foods.remove(food)
            foods.append(spawn_food())
            continue
        pygame.draw.rect(screen, BLUE, [food["x"], food["y"], 15, 15])
        timer_text = font.render(str(time_left // 1000 + 1), True, RED)
        screen.blit(timer_text, (food["x"], food["y"] - 20))

        head_x, head_y = snake[-1]
        if abs(head_x - food["x"]) < 15 and abs(head_y - food["y"]) < 15:
            scope += 1
            foods.remove(food)
            foods.append(spawn_food())
            
    score_text = font.render(f"SCORE: {scope}", True, WHITE)
    screen.blit(score_text, (10, 10))

    pygame.display.update()
    clock.tick(10)

cursor.execute("INSERT INTO user_score (user_id, level, score) VALUES (%s, %s, %s)", (user_id, scope, scope))
conn.commit()

pygame.quit()
```
