```python
import pygame
import datetime

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Mickey Mouse Clock")

bg = pygame.image.load('C:/Users/User/Downloads/mmbody.jpg')
bg = pygame.transform.scale(bg, (WIDTH, HEIGHT))

min_hand = pygame.image.load('C:/Users/User/Downloads/lefthand.png')   
sec_hand = pygame.image.load('C:/Users/User/Downloads/righthand.png')  

sec_hand = pygame.transform.scale(sec_hand, (800, 500))  
min_hand = pygame.transform.scale(min_hand, (40, 550))  

CENTER_X, CENTER_Y = 400, 250

def rotate_hand(surface, image, angle, pivot):
    rotated_image = pygame.transform.rotate(image, angle)
    rotated_rect = rotated_image.get_rect(center=pivot)
    surface.blit(rotated_image, rotated_rect)

clock = pygame.time.Clock()
running = True

while running:
    screen.blit(bg, (0, 0))  

    current_time = datetime.datetime.now()

    sec_angle = -current_time.second * 6  
    min_angle = -current_time.minute * 6  

    rotate_hand(screen, sec_hand, sec_angle, (CENTER_X, CENTER_Y + 50))  
    rotate_hand(screen, min_hand, min_angle, (CENTER_X, CENTER_Y + 50))  

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    pygame.display.update()
    clock.tick(30)

pygame.quit()
```

```python
import pygame
import os

pygame.init()
pygame.mixer.init()

screen = pygame.display.set_mode((500, 300))
pygame.display.set_caption("Music Player")

MUSIC_FOLDER = "C:/Users/User/Music/"  
tracks = [f for f in os.listdir(MUSIC_FOLDER) if f.endswith('.mp3')]

if not tracks:
    print("Нет MP3 файлов в указанной папке!")
    pygame.quit()
    exit()

current_track = 0
playing = False  

def play_music():
    global playing
    pygame.mixer.music.load(os.path.join(MUSIC_FOLDER, tracks[current_track]))
    pygame.mixer.music.play()
    playing = True
    print(f"Playing: {tracks[current_track]}")

def stop_music():
    global playing
    pygame.mixer.music.stop()
    playing = False
    print("Stopped")

def next_track():
    global current_track
    current_track = (current_track + 1) % len(tracks)
    play_music()
    print("Next Track")

def prev_track():
    global current_track
    current_track = (current_track - 1) % len(tracks)
    play_music()
    print("Previous Track")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_p:  # Play/Pause
                if playing:
                    pygame.mixer.music.pause()
                    playing = False
                    print("Paused")
                else:
                    pygame.mixer.music.unpause()
                    playing = True
                    print("Resumed")
            elif event.key == pygame.K_s:  # Stop
                stop_music()
            elif event.key == pygame.K_n:  # Next
                next_track()
            elif event.key == pygame.K_b:  # Previous
                prev_track()

pygame.quit()
```

```python
import pygame

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Moving Ball")

# Цвета
WHITE = (255, 255, 255)
RED = (255, 0, 0)

RADIUS = 25
x, y = WIDTH // 2, HEIGHT // 2
SPEED = 20  
running = True
while running:
    pygame.time.delay(50)  
    screen.fill(WHITE)  

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT] and x - RADIUS - SPEED >= 0:
        x -= SPEED
    if keys[pygame.K_RIGHT] and x + RADIUS + SPEED <= WIDTH:
        x += SPEED
    if keys[pygame.K_UP] and y - RADIUS - SPEED >= 0:
        y -= SPEED
    if keys[pygame.K_DOWN] and y + RADIUS + SPEED <= HEIGHT:
        y += SPEED

    pygame.draw.circle(screen, RED, (x, y), RADIUS)

    pygame.display.update()

pygame.quit()
```
